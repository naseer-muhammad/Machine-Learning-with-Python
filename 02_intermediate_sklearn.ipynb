## 1. Simple Linear Regression

### Introduction
ALL THESE TOPICS ARE THE PART OF SUPERVISED LEARNING


Below we will build a simple linear regression model to predict the tip amount based on the total bill amount using the `tips` dataset from seaborn. 

It is a `Regression model` because we are predicting the numerical values (tip amount) based on another numerical value (total bill amount) and the relationship between the two can be modeled with a linear equation so it is `linear`.

```python
# Let's write the whole code in one cell
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

# load the data
df = sns.load_dataset('tips')
# select the feature and target columns
X = df[['total_bill']]
y = df['tip']
# load/call the linear regression model
model = LinearRegression()
# fit the model means train the model
model.fit(X, y)
# predict the tip for a total bill of $35
model.predict([[35]])  # Predict tip for a total bill of $30
```

```python
sns.lmplot(x='total_bill', y='tip', data=df)
```

```python
# Let's write the whole code in one cell
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

# load the data
df = sns.load_dataset('tips')
# select the feature and target columns
X = df[['total_bill', 'size']]
y = df['tip']
# load/call the linear regression model
model = LinearRegression()
# fit the model means train the model
model.fit(X, y)
# predict the tip for a total bill of $35
model.predict([[35, 2]])  # Predict tip for a total bill of $35 and size of 2
```

### 2.1. Encoding Non-numerical "Smokers"

```python
# Let's write the whole code in one cell
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

# load the data
df = sns.load_dataset('tips')
# To include the smoker, we need to convert it into numerical values
df['smoker_numeric'] = df['smoker'].map({'Yes': 1, 'No': 0})
# select the feature and target columns
X = df[['total_bill', 'size', 'smoker_numeric']]
y = df['tip']
# load/call the linear regression model
model = LinearRegression()
# fit the model means train the model
model.fit(X, y)
# predict the tip for a total bill of $35, size of 2, and smoker

```

```python
print(model.predict([[50, 2, 0]]))  # Predict tip for a total bill of $50, size of 2, and non-smoker
print(model.predict([[50, 2, 1]]))  # Predict tip for a total bill of $50, size of 2, and smoker



# The variable of smoker is not a good predictor for tip amount in this case.
```

# 3. Classification

## 3.1. Logistic Regression
Its `classification` because we are predicting `categorical values` (smoker or non-smoker) based on numerical values (total bill amount, size of the party, and tip amount).

### 3.1.1. Binary Classification

```python
# Let's write the whole code in one cell
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression, LogisticRegression


# load the data
df = sns.load_dataset('tips')
# select the feature and target columns
X = df[['total_bill', 'size', 'tip']]
y = df['smoker']
# load/call the logistic regression model
model = LogisticRegression()
# fit the model means train the model
model.fit(X, y)
# predict the tip for a total bill of $35, size of 2, and smoker
```

```python
#predict 
model.predict([[45, 2, 5.67]])  # Predict smoker status for a total bill of $45, size of 2, and tip of $5.67
```

# 5. Regression Metrics

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression

#load the data
df = sns.load_dataset('tips')
df.head()



X = df[['total_bill', 'size']]
y = df['tip']

# train test split
from sklearn.model_selection import train_test_split
# X_train, X_test, y_train, y_train = train_test_split(X, y, test_size = 0.2)
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, random_state=1)


# call the model
model = LinearRegression()
# fit the model
model.fit(X_train, y_train)

# metric to evaluate the model
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, mean_absolute_percentage_error
y_pred = model.predict(X_test)


print(f"Mean Squared Error: {mean_squared_error(y_test, y_pred)}")
print(f"R2 Score: {r2_score(y_test, y_pred)}")
print(f"Mean Absolute Error: {mean_absolute_error(y_test, y_pred)}")
print(f"Mean Absolute Percentage Error: {mean_absolute_percentage_error(y_test, y_pred)}")
# root mean squared error
print(f"Root Mean Squared Error: {np.sqrt(mean_squared_error(y_test, y_pred))}")
```
