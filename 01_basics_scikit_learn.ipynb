# 1. Scikit Learn Library Basic Usage

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Simple linear regression
from sklearn.linear_model import LinearRegression
```

```python
# load the data
df = sns.load_dataset('tips')

# or we can also use the pd.read_csv() function to load the data from a csv file
# df = pd.read_csv('path_to_file.csv')

# display the first 5 rows of the dataframe
df.head()
```

```python
# create a regression line plot using seaborn
sns.lmplot(x='total_bill', y='tip', data=df)
```

```python
# select the feature column from the dataframe df
X = df[['total_bill']]
# select the target column from the dataframe df
y = df['tip']
```

```python
# load/call the model linear regression
model = LinearRegression()
# fit the model to the data X and y (fitting the model means training the model)
model.fit(X, y)
```

```python
# Let's predict the tip for an unknown total_bill amount
model.predict([[155]])  # Predict tip for a total bill of $50
```

```python
# Let's write the whole code in one cell

# load the data
df = sns.load_dataset('tips')
# select the feature and target columns
X = df[['total_bill']]
y = df['tip']
# load/call the linear regression model
model = LinearRegression()
# fit the model
model.fit(X, y)
# predict the tip for a total bill of $40
model.predict([[40]])  # Predict tip for a total bill of $40
```
