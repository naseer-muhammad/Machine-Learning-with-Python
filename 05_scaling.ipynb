## 1. Standard Scaling

`Standard scaling`(Z-fromula) is a method of scaling the data such that the distribution of the data is centered around $0$, with a standard deviation of $1 (-1$ to $1)$. This is done by subtracting the mean of the data from each data point and then dividing by the standard deviation of the data. This is a very common method of scaling data, and is used in many machine learning algorithms.

In Standard Scaling, the data can range between $-3$ to $3$.

The formula is as follows:
$$ z = \frac{x - \mu}{\sigma} $$
where $\mu$ is the `mean` of the data, and $\sigma$ is the `standard deviation` of the data.

```python
# import the libraries
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler, MaxAbsScaler
```

```python
# make an example dataset
df = {'Age': [25, 30, 35, 40, 45],
       'Height': [165, 170,175,180,185],
         'Weight': [55,60,65,70,75]}
df
```

```python
# Convert the above data to pandas dataframe
df = pd.DataFrame(df)
df
```

```python
# import the scalar
scalar = StandardScaler()
scalar
```

```python
# fit the scalar on data 
scaled_df = scalar.fit_transform(df)
scaled_df
```

```python
# Now convert this data into a pandas dataframe
scaled_df = pd.DataFrame(scaled_df, columns=df.columns)
scaled_df
```

## 2. Min-Max Scaling
In Min-Max scaling, the data is scaled to a range of $0$ to $1$. This is done by subtracting the minimum value of the data from each data point and then dividing by the range of the data. The formula is as follows:

$$ X_{scaled} = \frac{X - X_{min}}{X_{max}-X_{min}} $$
Where:
- $X$ is the original data point. e.g. $50$
- $X_{min}$  is the minimum value in the data.
- $X_{max}$ is the maximum value in the data.
- $(X_{scaled})$ is the resulting scaled value.

e.g. Consider a set of values with a minimum of $10$ and a maximum of $100$.
$ X_{scaled} = \frac{50 - 10}{100-10} = \frac {40}{90} = 0.4 $




```python
# import the scalar
scalar = MinMaxScaler()
scalar
```

```python
# fit the scalar on data
scaled_df = scalar.fit_transform(df)
scaled_df
```

```python
# Convert this data into a pandas dataframe
scaled_df = pd.DataFrame(scaled_df, columns=df.columns)
scaled_df
```

## 3. Max-Abs Scaler
In MaxAbs scaling, the data is scaled to a range of $-1$ to $1$. This is done by dividing the data by the maximum absolute value of the data. The formula is as follows:

$$ X_{scaled} = \frac{X}{\\max(|X|)} $$
Where: 
- $X$ is the original data point (or value for a specific feature/column).
- $max(|X|)$ is the maximum absolute value of all data points in that feature/column.
- $(X_{scaled})$ is the resulting scaled value. 

```python
# Import the scalar
scalar = MaxAbsScaler()
scalar
```

```python
# fit the scalar on data
scaled_df = scalar.fit_transform(df)
scaled_df
```

```python
# Convert this data into a pandas dataframe
scaled_df = pd.DataFrame(scaled_df, columns=df.columns)
scaled_df
```

## 5. Robust Scaling
In Robust Scaling, the data is scaled to a range of $-1$ to $1$. This is done by subtracting the median of the data from each data point and then dividing by the interquartile range of the data. The formula is as follows:


$$X_{scaled}=\frac{X - \text{Median} (X)}{\text{IQR}(X)}$$

Where: 
- $X$ is the original value of the feature.
- $\text{Median}(X)$ is the median (50th percentile or Q2) of the feature's values in the training data.
- $\text{IQR}(X)$ is the Interquartile Range, which is the difference between the 3rd quartile (Q3, 75th percentile) and the 1st quartile (Q1, 25th percentile): $\text{IQR}(X)=Q3(X)-Q1(X)$

```python
from sklearn.preprocessing import RobustScaler
# import the scalar
scalar = RobustScaler()
scalar
```

```python
# fit the scalar on data
scaled_df = scalar.fit_transform(df)
scaled_df
```

```python
# Convert this data into a pandas dataframe
scaled_df = pd.DataFrame(scaled_df, columns=df.columns)
scaled_df
```

## 6. Data Transformations

Converting one type of data distribution into another type of data distribution is called `Data Transformation`. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# generate non-normal data (exponential Distribution)
np.random.seed(0) #here seed(0) means it will generate same data every time
df = np.random.exponential(size=1000, scale=1) #here size=number of data points, scale=standard deviation
df = pd.DataFrame(df, columns=['values'])   # here we are creating a dataframe with column name 'values'
df
```

```python
df.head()
```

```python
# plot the histogram of the expoential data using seaborn
sns.histplot(df['values'], kde=True) # here we plot the column 'values' and we use kde=True to plot the kernel density estimate
# if we use the kde=False, we will not get the line passing through the histogram
```

```python
# Now we will transform the above expoenntial data into different distributions
from sklearn.preprocessing import PowerTransformer
from sklearn.preprocessing import QuantileTransformer

pt_boxcox = PowerTransformer(method='box-cox', standardize=False)
# here we are using box-cox method to transform the data into normal distribution, standardize=False means we will not standardize the data
pt_yeo_johnson = PowerTransformer(method='yeo-johnson', standardize=False)
# here we are using yeo-johnson method to transform the data into normal distribution, standardize=False means we will not standardize the data
qt_normal = QuantileTransformer(output_distribution='normal')
# here we are using the QuantileTransformer method by keeping the output_distribution as normal
```

```python
# for box-cox, the data must be positive 
df['Box Cox'] = pt_boxcox.fit_transform(df[['values']] +1) # here we are adding 1 to the data to make it positive
df['Yeo Johnson'] = pt_yeo_johnson.fit_transform(df[['values']])
df['Quantline'] = qt_normal.fit_transform(df[['values']])
```

```python
df
```

```python
df.head()
```

```python
# Create histograms for all the coulumn using sns.hist and kde=True 
# Use a for loop
for col in df:
    sns.histplot(df[col], kde=True)
    plt.title(col)
    plt.show()
```

## 7. Normalization
Normalization is a broad term for making something more regular or standard.

`Data Normalization` is a technique often applied as part of data preparation for machine learning. The goal of normalization is to change the values of `numeric columns` in the dataset to a common scale, without distorting differences in the ranges of values. 

For machine learning, every dataset does not require normalization. It is required only when features have different ranges.

### 7.1. $L2$ Normalization
Rescales each sample (row) to have unit norm. This type of normalization is often used when dealing with text data. The L2 norm is calculated as the square root of the sum of the squared vector values.

The L2 normalized vector is given by $\mathbf{x}' = \frac{\mathbf{x}}{\|\mathbf{x}\|_2}$.
where 
- $\mathbf{x}$ is the original vector, and 
- $\|\mathbf{x}\|_2$ is the L2 norm of the vector.

e.g. for vector $x$ = [3, 4, 2]

$\|\mathbf{x}\|_2= \sqrt{3^2 + 4^2 + 2^2} = \sqrt{29}=5.39$
then $$x' = \frac{x}{\|\mathbf{x}\|_2} = \frac{[3, 4, 2]}{5.39} = [0.56, 0.74, 0.37]$$




```python
from sklearn.preprocessing import Normalizer
data = [[1, 1, 1], [1, 1, 0], [1, 0, 0]]
print(data)
```

```python
normalizer = Normalizer(norm='l2') # The 'norm = l2' 
print(normalizer.fit_transform(data))
```

### 7.2. $L1$ Normalization:
Also rescales each sample (row) but with a different approach, ensuring the sum of the absolute values is 1 in each row. The `L1` norm is calculated as the sum of the absolute vector values. Example:

```python
from sklearn.preprocessing import Normalizer
data = [[1, 1, 1], [1, 1, 0], [1, 0, 0]]
normalizer = Normalizer(norm='l1')
print(normalizer.fit_transform(data))
```

## 8. Log Transformation
In Log Transformation, we take the natural logarithm of the data. The formula is as follows:

$$ X_{scaled} = log(X) $$
Where:
- $X$ is the original data point. e.g. $50$
- $X_{scaled}$ is the resulting scaled value. e.g. $log(50) = 1.6094379124341003$

```python
import pandas as pd
import numpy as np

# example dataset with skewed values
df = { "Values": [1,5,10,20,50,100,200,500,1000,2000,5000,10000,20000,50000,100000,1000000]}
df = pd.DataFrame(df)
df
```

```python
df['log_values'] = np.log(df['Values'])
df
```

```python
for col in df:
    sns.histplot(df[col], kde=True)
    plt.title(col)
    plt.show()
```

In this way, we can see that the distribution of the data has become more normal instead of skewed as in the original values.
